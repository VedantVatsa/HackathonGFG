```python
import pandas as pd
import matplotlib.pyplot as plt
import io

```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['business_data.csv']))
```

#Stocks Holdings


```python
line, = plt.plot(df.iloc[:,0], df.iloc[:,1])
line.set_color("red")
plt.xlabel('Year')
plt.ylabel('Value of stocks holding that year')
plt.title('Plot of Stocks holding in the recent FYs')
plt.show()
```

#Short Term Investment

```python
plt.plot(df.iloc[:,0], df.iloc[:,2])
plt.xlabel('Year')
plt.ylabel('Value of net short term investments')
plt.title('Plot of Short Term Investments')
plt.show()
```

#Assets

```python
#For Assets
x = df.iloc[:,0]

plt.plot(x, df.iloc[:,1], label="Stocks")
plt.plot(x, df.iloc[:,2], label="STI")
plt.plot(x, df.iloc[:,3], label="LTI")
plt.plot(x, df.iloc[:,4], label="Profit")
plt.plot(x, df.iloc[:,5], label="Property")


plt.xlabel('Financial Year')
plt.ylabel('Assets')
plt.title('Plot of all assets in the recent FYs')
plt.legend()
plt.show()
```

#Liabilities

```python
#For Liabilities
x = df.iloc[:,0]

plt.plot(x, df.iloc[:,6], label="Loan")
plt.plot(x, df.iloc[:,7], label="Expenditure")
plt.plot(x, df.iloc[:,8], label="Tax-Paid")
plt.plot(x, df.iloc[:,9], label="Daily-Living-Expenses")


plt.xlabel('Financial Year')
plt.ylabel('Liabilities')
plt.title('Plot of all liabilities in the recent FYs')
plt.legend()
plt.show()
```

#Assets and Liabilities

```python
df['Total Investments'] = df.iloc[:,1] + df.iloc[:,2] + df.iloc[:,3] + df.iloc[:,4] + df.iloc[:,5]
df['Total Liability'] = df.iloc[:,6] + df.iloc[:,7] + df.iloc[:,8] + df.iloc[:,9]
plt.plot(x, df.iloc[:,10], label="Total-Investments")
plt.plot(x, df.iloc[:,11], label="Total-Liablities")
plt.xlabel("Financial Year")
plt.ylabel("Assets and Liabilities")
plt.title("Plot of assets and liablities in the recent FYs")
plt.legend()
plt.show()
```

#Net Worth

```python
df['Net-Worth'] = df.iloc[:,10] - df.iloc[:,11]
line, = plt.plot(x, df.iloc[:,12], label="Net-Worth")
line.set_color("red")
plt.xlabel("Financial Year")
plt.ylabel("Net Worth")
plt.title("Plot of net worth in the recent FYs")
plt.legend()
plt.show()
```
